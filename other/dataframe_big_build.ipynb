```python
import pandas as pd
import numpy as np
from dataset import *
from save_load import *
from NN_library.AutoUnet.AutoUnet_uniform import *
from NN_library.ConvNet.ConvNet_uniform import *
%load_ext autoreload
%autoreload 2
```

```python
unet = AutoUNet()
convnet = ConvNet(16)

args = {'lr' : 0.001, 'epochs' : 20, 'dev' : dev, 'name' : 'NN_library/ConvNet/Conv_w_AutoUnet'}
convnet = load_network(convnet, 'NN_library/ConvNet/ConvNet_uniform_49', args)
unet = load_network(unet, 'NN_library/AutoUnet/AutoUnet_uniform_27', args)  

unet.to(args['dev'])
convnet.to(args['dev'])
```

```python
data = dataset_big()
loader = DataLoader(data, batch_size = 32, shuffle=False, num_workers=0)
```

```python
damage = []
shrinkage = []
stiffness_1 = []
stiffness_2 = []
interface = []
```

```python
for i, (geometry, imp_shrinkage) in enumerate(loader):
    geometry = geometry.to(args['dev'])
    imp_shrinkage = imp_shrinkage.to(args['dev']) / -0.001
    interface += torch.where(geometry==0.5, 1, 0).sum(axis=(1,2,3)).tolist()
    for n in range(10):
        if n == 0:
            x = torch.cat([geometry, imp_shrinkage[:,[1],:,:], imp_shrinkage[:,[0],:,:]], axis=1)
        else:
            x = torch.cat([geometry, imp_shrinkage[:,[n+1],:,:], y.detach()], axis=1)
        y = unet(x)
        x_shr = torch.cat([geometry, imp_shrinkage[:,[n+1],:,:], y.detach()], axis=1)
        y_shr = convnet(x_shr)
        if n == 0:
            stiffness_1 += y_shr[:,0].tolist()
        if n == 9:
            stiffness_2 += y_shr[:,0].tolist()
            damage += (y.sum(axis=(1,2,3)) / 99**2).tolist()
            shrinkage += y_shr[:,1].tolist()
```
